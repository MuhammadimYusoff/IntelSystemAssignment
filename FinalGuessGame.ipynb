```python
#? Guessing Game/Word Jumbler Game
from tkinter import *
from random import choice, shuffle

root = Tk()
root.title('Guess The Word? Name of the Almighty')

myLabel = Label(root, text="", font=("Helveltica", 30))
myLabel.pack()

def shuffler():
    # Clear Hint Label and Hint Count
    hintLabel.config(text='')
    global hintCount
    hintCount = 0

    # Clear Previous Answer and Answer Label
    entryAnswer.delete(0, END)
    answerLabel.config(text='')

    # List of Names words
    names = ["ALADL", "ALMALIK", "ASSALAM", "ALKHALIQ", "ALKAREEM"]

    # Pick Random Name from List
    global word
    word = choice(names)
    myLabel.config(text=word)

    # Break apart chosen word
    breakApartWord = list(word)
    shuffle(breakApartWord)
    
    # Turn shuffled list into a word
    global shuffledWord
    shuffledWord = ""

    for letter in breakApartWord:
        shuffledWord += letter
    
    # Print shuffled word to the screen
    myLabel.config(text=shuffledWord)

global FN # f(n) = g(n) + h(n)
FN = 0

global HN # Number of Wrong Letters Position
HN = 0

global GN # Number of Tries
GN = 0

global POINTS # Player's Final Score
POINTS = 0

# Answer function: Check for Correct Letter Position
def answer():
    global HN
    HN = 0
    
    global G
    G = GN
    G += 1

    global points
    points = POINTS

    userInputArray = [letter for letter in entryAnswer.get()]
    print(userInputArray)

    # Split the word to array
    splitWord = []
    for i in word:
        splitWord.append(i)
        print(splitWord) # Test to print in console

# 
    if word == entryAnswer.get():
        wrongLetters.config(text='')
        answerLabel.config(text="Correct Answer")
        points += 1 # Player's Score +1
    else:
        wrongLetters.config(text='')
        answerLabel.config(text="Incorrect Answer")
        
        HN = 0
        # Compare the Word Array with the User Input
        for i in range(len(userInputArray)):
            if userInputArray[i] != splitWord[i]:
                HN += 1
                print(HN)

        # Label of Wrong Letters position
        wrongLetters.config(text="Wrong Letters: " + str(HN))
        wrongLetters.pack(pady=10)


# Create Hint Counter
global hintCount
hintCount = 0

# Hint Function
def hint(count):
    if word == "ALADL":
        hintLabel.config(text="HINT: The Name means: The Justice \n")
    if word == "ALMALIK":
        hintLabel.config(text="HINT: The Name means: The King of Kings \n")
    if word == "ASSALAM":
        hintLabel.config(text="HINT: The Name means: Peace \n")
    if word == "ALKHALIQ":
        hintLabel.config(text="HINT: The Name means: The Creator \n")
    if word == "ALKAREEM":
        hintLabel.config(text="HINT: The Name means: The Munificent \n")
    
    # Count the number of Hint letters and printout the letters if requested
    global hintCount
    hintCount = count
    wordLength = len(word) # The Word we will give hint

    # Printout letters for Hint
    if count < wordLength:
        hintLetters.config(text=f'{hintLetters["text"]} {word[count]}')
        hintCount += 1



############################### USER INTERFACE PART #########################################
entryAnswer = Entry(root, font=("Helveltica", 24))
entryAnswer.pack(pady=20)

# Shuffle the word
buttonFrame = Frame(root)
buttonFrame.pack(pady=20)

# Answer Button
answerButton = Button(buttonFrame, text="Answer", command=answer)
answerButton.grid(row=0, column=0, padx=10)

# Other Word Button
myButton = Button(buttonFrame, text="Pick Another Word", command=shuffler)
myButton.grid(row=0, column=1, padx=10)

# Hint Button
hintButton = Button(buttonFrame, text="Hint",command=lambda: hint(hintCount))
hintButton.grid(row=0, column=2, padx=10)

# Answer Label
answerLabel = Label(root, text='',font=("Helveltica", 18))
answerLabel.pack(pady=20)

# Points Label
pointsLabel = Label(root, text= "Score: " + str(POINTS) ,font=("Helveltica", 18))
pointsLabel.pack(pady=20)

# Label to show hint and letters if hint button pressed
hintLabel = Label(root, text='', font=("Helveltica", 18))
hintLabel.pack(pady=10)
hintLetters = Label(root, text='', font=("Helveltica", 18))
hintLetters.pack(pady=10)

# Display number of wrong letters position
wrongLetters = Label(root, text='', font=("Helveltica", 18))
wrongLetters.pack(pady=10)

# Run the Shuffler on start
shuffler()
root.mainloop()
```

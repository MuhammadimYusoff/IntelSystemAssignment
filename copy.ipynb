```python
#attempt 2 (in progress) §↔§
import math 
import numpy as np 

name = np.array([['A', 'L', 'A', 'D', 'L'], 
                 ['A', 'L', 'M', 'A', 'L', 'I', 'K'],
                 ['A', 'S', 'S', 'A', 'L', 'A', 'M'],
                 ['A','L','K','H','A','L', 'I', 'Q'],
                 ['A','L','K','A','R','E','E','M']])
global gn, hn, fn, en
gn = 0 # no of tries
hn = 0 # wrong letters
fn = 0 # f(n) = g(n) + h(n)
en = 0 # no of entries
points = 0

def guess():
    gn += 1
    while True:
        userinput = input("Guess the word below!\n")
        en += 1
        if userinput == name[5]:
            break
        if userinput != name[5] and en > 5:
            break
        for element in name[0]:
            if element != userinput:
                hn += 1
                print(userinput)
        print("Wrong letters → "+ str(hn))
        print("Tries: "+ str(gn))
        fn = gn + hn
        print("f(n) "+ str(fn))

        if hn > 0:
            print("Try again")
        else:
            print("Correct! You got it.")
            points += 1
            print("Your score: "+ str(points))
```

```python
#? Guessing Game/Word Jumbler Game
from tkinter import *
from random import choice, shuffle

root = Tk()
root.title('Guess The Word? Name of the Almighty')

myLabel = Label(root, text="", font=("Helveltica", 30))
myLabel.pack()

def shuffler():
    # Clear Previous Answer and Answer Label
    entryAnswer.delete(0, END)
    answerLabel.config(text='')

    # List of States words
    # states = ["Malaysia", "Indonesia", "Egypt", "Bangladesh", "Iran", "Iraq", "Turkiye", "China"]
    names = ["AL ADL", "AL MALIK", "AS SALAM", "AL KHALIQ", "AL KAREEM"]

    # Pick Random State from List
    global word
    # word = choice(states)
    word = choice(names)
    myLabel.config(text=word)

    # Break apart chosen word
    breakApartWord = list(word)
    shuffle(breakApartWord)
    
    # Turn shuffled list into a word
    global shuffledWord
    shuffledWord = ""

    for letter in breakApartWord:
        shuffledWord += letter
    
    # Print shuffled word to the screen
    myLabel.config(text=shuffledWord)

global FN
FN = 0

global HN
HN = 0

global GN
GN = 0

global POINTS
POINTS = 0

# Answer function
def answer():
    global G
    G += 1
    if word == entryAnswer.get():
        answerLabel.config(text="Correct Answer")
        POINTS += 1
    else:
        answerLabel.config(text="Incorrect Answer")
        HN += 1

# Create Hint Counter
global hintCount
hintCount = 0

# Hint Function
def hint(count):
    global hintCount
    hintCount = count
    # The Word we will give hint
    wordLength = len(word)

    # Show Hint
    if count < wordLength:
        hintLabel.config(text=f'{hintLabel["text"]} {word[count]}')
        hintCount += 1

entryAnswer = Entry(root, font=("Helveltica", 24))
entryAnswer.pack(pady=20)

# Shuffle the word
buttonFrame = Frame(root)
buttonFrame.pack(pady=20)
myButton = Button(buttonFrame, text="Pick Another Word", command=shuffler)
myButton.grid(row=0, column=0, padx=10)

# Answer Button
answerButton = Button(buttonFrame, text="Answer", command=answer)
answerButton.grid(row=0, column=1, padx=10)

# Hint Button
hintButton = Button(buttonFrame, text="Hint",command=lambda: hint(hintCount))
hintButton.grid(row=0, column=2, padx=10)

# Answer Label
answerLabel = Label(root, text='',font=("Helveltica", 18))
answerLabel.pack(pady=20)

# Answer POINTS
POINTSLabel = Label(root, text= POINTS ,font=("Helveltica", 18))
POINTSLabel.pack(pady=20)

# Label to show hint if hint button pressed
hintLabel = Label(root, text='', font=("Helveltica", 18))
hintLabel.pack(pady=10)

# Run the Shuffler on start
shuffler()
root.mainloop()
```
